```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_palette('hls')
```

```python
sns.countplot(df['Revisited_point'])

plt.title( "Revisited Points Density" , size = 21 )
plt.xlabel( "Revisited Point" , size = 11 )
plt.ylabel( "Density" , size = 11 )
```

```python
df['CaCO3'].value_counts().plot(kind='pie',autopct='%.2f')

plt.title( "CaCO3 Pie Chart" , size = 21 )
```

```python
plt.hist(df['Clay'],bins=11)

plt.title( "Clay Density" , size = 21 )
plt.xlabel( "Clay" , size = 11 )
plt.ylabel( "Density" , size = 11 )
```

```python
sns.distplot(df['Clay'])

plt.title( "Clay Density" , size = 21 )
plt.xlabel( "Clay" , size = 11 )
plt.ylabel( "Density" , size = 11 )
```

```python
sns.scatterplot(data=df,x='Clay',y='Coarse')

plt.title( "Clay vs Coarsity" , size = 21 )
plt.xlabel( "Clay" , size = 11 )
plt.ylabel( "Coarse" , size = 11 )
```

```python
sns.distplot(df['Clay'],hist=False)
sns.distplot(df['Coarse'],hist=False)

plt.title( "Clay vs Coarsity" , size = 21 )
plt.xlabel( "Clay" , size = 11 )
plt.ylabel( "Coarse" , size = 11 )
```
